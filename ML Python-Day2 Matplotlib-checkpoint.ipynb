### Section1:Basic Line Plotting 

```python
#imports needed for functions 
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from pylab import *
```

```python
x = np.linspace(0, 5, 10) #10 samples points from 0 to 5
print(x) #will print the array
print(type(x)) #numpy array
y = x**2 #the equation y is x raised to 2
print(y)
```

```python
plt.figure()
```

```python
plt.plot(x, y, 'r') #plot x and y and 'r' - red 
#xlabel('x')
plt.xlabel('timesample')
#ylabel('y')
plt.ylabel('temperature values')
plt.title('my first graph')
# plt.grid()
plt.show() #mandatory without this won't show output
```

```python
plt.plot(x, y, 'r', label = 'temp 1')
plt.plot(y, x, 'g',label = 'temp 2') 
#xlabel('x')
plt.xlabel('timesample')
#ylabel('y')
plt.ylabel('temperature values')
plt.title('my first graph')
plt.legend()
plt.grid()

plt.show()
```

## Subplotting

```python
#subplot
plt.subplot(1,2,1) #1 row 2 columns 1st one
plt.plot(x, y, 'r--') # y w.r.t x and r-- dashed line in red
plt.subplot(1,2,2) #1 row 2 columns 2nd one
plt.plot(y, x, 'g*-') #x w.r.t. y *- style green line
plt.show()
```

```python
x1 = np.linspace(5,50,10)
y1 = 3*x1
x2 = np.linspace(5,50,10)
y2 = 5*x2**2+7
plt.figure()
plt.subplot(2,2,1)
plt.plot(x1,y1,'m+')
plt.subplot(2,2,2)
plt.plot(y1,x1, 'b--')
plt.subplot(2,2,3)
plt.plot(x2,y2,'g:')
plt.subplot(2,2,4)
plt.plot(y2,x2,'y*-')
plt.show()
```
