Concat()

```python
import pandas as pd

# Creating two DataFrames
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
print (df1)
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})
print (df2)
# Using concat
result = pd.concat([df1, df2], ignore_index=True) #to mearge the two table/raws or columns
print(result)
```

```python
# concat by raws

import pandas as pd

df1 = pd.DataFrame({'Name': ['Alice', 'Bob'], 'Age': [25, 30]})
print (df1)
df2 = pd.DataFrame({'Name': ['Charlie', 'David'], 'Age': [35, 40]})
print (df2)
result = pd.concat([df1, df2], ignore_index=True)
print(result)

# output

#       Name     Age
# 0   Alice       25
# 1      Bob      30
# 2  Charlie      35
# 3    David      40
```

```python
# concat by column

df3 = pd.DataFrame({'City': ['Pune', 'Delhi']})
result = pd.concat([df2, df3], axis=1)
print(result)


# output
#        Name      Age    City
# 0     Alice      25     Pune
# 1       Bob      30     Delhi
```
