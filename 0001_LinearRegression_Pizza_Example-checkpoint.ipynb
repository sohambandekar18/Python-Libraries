## Single Variable Linear Regression

Simple linear regression can be used to model a linear relationship between one response variable and one explanatory variable

Predicting Pizza prices

Let's model the relationship between the size of a pizza and its price. First,
we will write a program with scikit-learn that can predict the price of a pizza given  its size. Then, we will discuss how simple linear regression works and how it can  be generalized to work with other types of problems. Let's assume that you have  recorded the diameters and prices of pizzas that you have previously eaten in your pizza journal

Training instance Diameter (in inches) Price (in dollars) 

| Record | Diameter(Inch) | Price$| 
|--------|----------------|-------| 
| 1      |  6             |  7    | 
| 2      |  8             |  9    | 
| 3      |  10            |  13   | 
| 4      |  14            |  17.5 | 
| 5      |  18            |  18   | 


```python
%matplotlib inline
import matplotlib.pyplot as plt

# List of List input and output format is important 
# Ordinary Least Square takes Multiple Input and Predicts Multiple Outputs

PSize = [[6], [8], [10], [14], [18]] #input has to lists of lists
PPrice = [7, 9, 13, 17.5, 18] #outputs is a single list

# import pandas as pd
# # Read the Excel file
# df = pd.read_csv("pizza_data.csv") #excel file saved in same folder (downloads)
# # Display the dataframe
# print(df)

plt.figure()
plt.title('Pizza price plotted against diameter')
plt.xlabel('Diameter in inches')
plt.ylabel('Price in dollars')
plt.plot(PSize, PPrice, 'go')
plt.axis([0, 25, 0, 25]) #xlims & ylims together
plt.grid(True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split # scikit learn -machine learning  and train/test/split
PSize_train, PSize_test, PPrice_train, PPrice_test = \
train_test_split(PSize, PPrice, test_size=0.2) # 20% data will go for testing, 80% data will go for 

plt.figure()
plt.title('Pizza price plotted against diameter')
plt.xlabel('Diameter in inches')
plt.ylabel('Price in dollars')
plt.plot(PSize_train, PPrice_train, 'go')
plt.axis([0, 25, 0, 25])
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

# Create and fit the model
model = LinearRegression()
model.fit(PSize_train, PPrice_train)

print(PPrice_test)                          ## Test output - Expected Output
model.predict(PSize_test)                   ## Observe Output its a single list
```

```python
# Plot Initial Data 
plt.figure()
plt.title('Pizza Price Prediction against diameter')
plt.xlabel('Diameter in inches')
plt.ylabel('Price in dollars')
plt.plot(PSize_train, PPrice_train, 'go')
plt.plot(PSize_test, PPrice_test, 'bo')

plt.axis([0, 25, 0, 25])
plt.grid()
plt.show()
```

```python

# Plot Regression Solution line
regression_line_y = model.predict([[0], [25]])
print(regression_line_y)
plt.plot([0, 25], regression_line_y, 
         color='r', linestyle='-', linewidth=1)
plt.show()
```

```python
print('A 9" pizza should cost: $%.2f' % model.predict([[9]]))     # Prediction of inexperienced data
```

```python
model.score(PSize, PPrice)
```

```python
print("Slope(m) : %.2f, Intercept (c): %.2f"%
      (model.coef_, model.intercept_))
```
