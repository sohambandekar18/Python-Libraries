```python
age = 16
if age < 13:
    # If age is less than 13, this block runs
    print("You are a child.")
elif age < 20:
    # If age is 13 or more but less than 20, this block runs
    print("You are a teenager.")
else:
    # If age is 60 or more, this block runs
    print("You are a senior.")
```

set

```python
# Create a set
my_set = {10, 20, 30} 
print("Original Set:", my_set)

# Add an element
my_set.add(40)
print("After adding 40:", my_set)

my_set.add(60)
print(my_set)

# Remove an element (will give error if not present)
# my_set.remove(100)
# print("After removing 20:", my_set)
my_set.discard(100)
print (my_set)

my_set.pop()
print("After popping:", my_set)

my_set.pop()
print("After popping:", my_set)

my_set.clear()
print("After clearing:", my_set)
```

```python
# Discard an element (no error if not present)
my_set.discard(100)
 
# 100 is not in the set
print("After discarding 100 (no error):", my_set)

# Pop a random element
my_set.pop()
print("After popping:", my_set)

# Clear the set
my_set.clear()
print("After clearing:", my_set)
```

List

```python
fruits = ["apple", "banana", "cherry"]
fruits.append("mango")  # Add item
print (fruits)
print(fruits[1])        # Access item
```

```python
# Initial list
fruits = ["apple", "banana", "cherry"]
print("Original List:", fruits)

# append(item) 
fruits.append("mango")
print("After append:", fruits)

# insert(index, item) 
fruits.insert(1, "arange")
print("After insert at index 1:", fruits)
```

```python
# remove(item) 
fruits.remove("banana")
print("After remove 'banana':", fruits)

# pop(index) 
fruits.pop()  # removes last item
print("After pop (last item removed):", fruits)

# sort() 
fruits.sort()
print("After sort:", fruits)

# reverse() 
fruits.reverse()
print("After reverse:", fruits)
```
