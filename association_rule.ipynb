```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori

dataset = [["Bread", "Milk", "Beer"],
            ["Bread", "Diapers", "Eggs"],
            ["Milk", "Diapers", "Beer", "Cola"],
            ["Bread", "Milk", "Diapers", "Beer"],
            ["Bread", "Milk", "Cola"]] # the datatype is List of Lists
```

```python
te = TransactionEncoder()
te = te.fit(dataset)                    # Reads all unique items from the dataset
te_ary = te.transform(dataset)          # Converts array into hot coded matric
print(te.columns_)           # transaction = column an draw wise structure
print(te_ary)
```

```python
df = pd.DataFrame(te_ary, columns=te.columns_)     # Panda dataframe
print(df.columns)
df
```

```python
# Finds the freuency of the item
frequent_itemsets = apriori(df, min_support=0.6, 
                            use_colnames=True) 
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules
rules = association_rules(frequent_itemsets, 
                          metric="support", 
                          min_threshold=0.5)
rules
```

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(frequent_itemsets, 
                          metric="confidence", 
                           min_threshold=0.7)
rules
```

```python
rules[rules['support'] == rules['support'].max()]
```
